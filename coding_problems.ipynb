Find find_longest_chain length

```python
from collections import defaultdict, deque



"""

Aman shares a funny joke with his friends. Each person who listens to the joke 
finds it hilarious and shares it with their own set of friends. The impact of 
the joke is represented by a sequence of strings in the format(m number of columns) 
al listened where is the person who hears the joke from, has already heard it.
Aman, registered as "Aman," is the original source of the joke. The goal is to 
determine the popularity of Aman's joke by calculating the length of the longest 
chain of people who have heard and shared the joke Input Format The first line of 
the input represents n denoting the row number. 
â€¢ The following line represents m denoting the column number. 
(the number of columns in each event description) The next n lines 
represent the events that are listed as "namel listened name2" 
All the names are in lowercase English alphabets and can also be numbers.
"""
# Example usage:
n = 5  # Number of events
m = 3  # Number of columns (not used directly in logic)
events = [
    "alice listened aman",
    "bob listened alice",
    "charlie listened bob",
    "dave listened charlie",
    "eve listened dave",
    "shiv listened aman",
    "shankar listened shiv",
    "alice listened shankar",

]

def find_longest_chain(n, m, events, start):

    graph = {}
    for obj in events:
        obj = obj.split(" ")
        if obj[-1] not in graph:
            graph[obj[-1]]  = [obj[0]]
        else:
            graph[obj[-1]].append(obj[0])
    
    counted = set()

    def dfs(curr):
        if curr not in graph or curr in counted:
            return 0
        counted.add(curr)
        max_ = 0
        for n in graph[curr]:
            max_ = max(max_, 1+dfs(n))
        
        counted.remove(curr)
        
        return max_
    
    print(graph)
    return 1+dfs(start)

    

print(find_longest_chain(n, m, events, "aman"))

```
