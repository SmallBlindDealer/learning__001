```python
from collections import deque

def findShortestPath(start, end, rooms):
    # If the start and end rooms are the same, return the single room
    if start == end:
        return [start]

    # Queue to store the current path
    queue = deque([[start]])
    
    # Set to keep track of visited rooms
    visited = set([start])
    
    # BFS loop
    while queue:
        print(queue)
        # Get the current path
        path = queue.popleft()
        # Get the last room in the current path
        current_room = path[-1]
        
        # Explore neighbors of the current room
        for neighbor in rooms.get(current_room, []):
            if neighbor == end:
                # Found the destination, return the path
                return path + [neighbor]
            if neighbor not in visited:
                # Mark neighbor as visited and enqueue the new path
                visited.add(neighbor)
                queue.append(path + [neighbor])
    
    # If no path is found, return an empty array
    return []

# Example usage:
rooms = {
    'Entrance': ['Hall', 'Kitchen'],
    'Hall': ['Entrance', 'Bedroom', 'Bathroom'],
    'Kitchen': ['Entrance', 'Pantry'],
    'Bedroom': ['Hall', 'Bathroom'],
    'Bathroom': ['Hall', 'Bedroom'],
    'Pantry': ['Kitchen', 'Treasure'],
    'Treasure': ['Pantry']
}

start = 'Entrance'
end = 'Treasure'
print(findShortestPath(start, end, rooms))
```

```python
import heapq

def WeightedPath(strArr):
    num_nodes = int(strArr[0])  # Number of nodes
    nodes = strArr[1:num_nodes+1]  # Node list
    edges = strArr[num_nodes+1:]  # Edges with weights
    
    # Build the graph as an adjacency list
    graph = {node: {} for node in nodes}
    for edge in edges:
        node1, node2, weight = edge.split('|')
        graph[node1][node2] = int(weight)
        graph[node2][node1] = int(weight)  # Bidirectional graph
    
    start, end = nodes[0], nodes[-1]  # Start node and end node
    
    # Dijkstra's algorithm to find the shortest path
    def dijkstra(graph, start, end):
        q = [(0, start, [])]
        visited = set()
        while q:
            weight, current, path = heapq.heappop(q)
            if current in visited:
                continue
            visited.add(current)
            path = path + [current]
            if current==end:
                return path
            
            for next_, wt in graph[current].items():
                if next_ not in visited:
                    heapq.heappush(q, (wt+weight, next_, path))
        
        return None

    # Find shortest path
    result_path = dijkstra(graph, start, end)
    
    if result_path:
        return '-'.join(result_path)
    else:
        return "-1"

# Example usage
print(WeightedPath(["4", "A", "B", "C", "D", "A|B|2", "C|B|1", "C|D|3", "B|D|2"]))  # Output: A-B-D
print(WeightedPath(["4", "x", "y", "z", "w", "x|y|2", "y|z|1", "z|w|2"]))  # Output: -1

arr = ["4", "A", "B","C", "D", "A|B|1", "B|D|9", "B|C|3", "C|D|4"]
```

```python
import heapq

def WeightedPath(strArr):
    num_nodes = int(strArr[0])  # Number of nodes
    nodes = strArr[1:num_nodes+1]  # Node list
    edges = strArr[num_nodes+1:]  # Edges with weights
    
    # Build the graph as an adjacency list
    graph = {node: {} for node in nodes}
    for edge in edges:
        node1, node2, weight = edge.split('|')
        graph[node1][node2] = int(weight)
        graph[node2][node1] = int(weight)  # Bidirectional graph
    
    start, end = nodes[0], nodes[-1]  # Start node and end node
    
    # Dijkstra's algorithm to find the shortest path
    def dijkstra(graph, start, end):
        heap = [(0, start, [])]  # (current weight, current node, path taken)
        visited = set()
        
        while heap:
            weight, current, path = heapq.heappop(heap)
            
            if current in visited:
                continue
            visited.add(current)
            path = path + [current]
            
            if current == end:
                return path
            
            for neighbor, edge_weight in graph[current].items():
                if neighbor not in visited:
                    heapq.heappush(heap, (weight + edge_weight, neighbor, path))
            
        
        return None
    
    # Find shortest path
    result_path = dijkstra(graph, start, end)
    
    if result_path:
        return '-'.join(result_path)
    else:
        return "-1"

# Example usage
print(WeightedPath(["4", "A", "B", "C", "D", "A|B|2", "C|B|1", "C|D|3", "B|D|2"]))  # Output: A-B-D
print(WeightedPath(["4", "x", "y", "z", "w", "x|y|2", "y|z|1", "z|w|2"]))  # Output: -1
```
